```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\berka\desktop\blob\Normal_400x_88.jpg')

plt.imshow(image, cmap='gray')
plt.title('ORİGİNAL', 
    fontweight ="bold")
plt.show()


gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Gray', 
    fontweight ="bold")
plt.show()


(T, thresh) = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.title('thresh', 
    fontweight ="bold")
plt.show()



# print(np.max(image))




# blur = cv2.GaussianBlur(gray, (11,11), 0)
blur = cv2.medianBlur(gray, 13)
plt.imshow(blur, cmap='gray')
plt.title('Blur', 
    fontweight ="bold")
plt.show()


# define the alpha and beta
alpha = 1.25 # Contrast control
beta = 65 # Brightness control
cont = cv2.convertScaleAbs(blur, alpha=alpha, beta=beta)


plt.imshow(cont, cmap='gray')
plt.title('Contrast and Brightness', 
    fontweight ="bold")
plt.show()



# canny = cv2.Canny(blur, 1, 255, 3)
canny = cv2.Canny(cont, 150,15)

plt.imshow(canny, cmap='gray')
plt.title('Canny', 
    fontweight ="bold")
plt.show()


dilated = cv2.dilate(canny, (1,1), iterations = 2)
plt.imshow(dilated, cmap='gray')
plt.title('Dilate', 
    fontweight ="bold")
plt.show()



# RETR_EXTERNAL 
# RETR_LIST 
# RETR_CCOMP 
# RETR_TREE
# CHAIN_APPROX_SIMPLE
# CHAIN_APPROX_NONE
(cnt, heirarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.drawContours(rgb, cnt, -1, (0,255,0), 3)


plt.imshow(rgb)
plt.show()

print('cell: ', len(cnt))
plt.show()
```
